# 03. polynomial regression
### 공부 시간에 따른 시험 점수

```python
import numpy as np
import matplotlib.pyplot as pit
import pandas as pd
```

```python
dataset = pd.read_csv('PolynomialRegressionData.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

### 단순 선형회귀

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

### 데이터 시각화

```python
pit.scatter(x,y,color = 'blue')
pit.plot(x,reg.predict(x),color = 'green')
pit.title('score by hours')
pit.xlabel('hours')
pit.ylabel('score')
pit.show()
```

```python
reg.score(x,y)
```

## 다항회귀

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)# 2차
x_poly =poly_reg.fit_transform(x)
x_poly[:5] # x를 0승, 1승, 2승으로 변환
```

```python
poly_reg.get_feature_names_out()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_poly,y)#변환된 x와 y를 가지고 모델 생성
```

### 데이터 시각화

```python
pit.scatter(x,y,color = 'blue')
pit.plot(x,lin_reg.predict(poly_reg.fit_transform(x)),color = 'green')
pit.title('score by hours')
pit.xlabel('hours')
pit.ylabel('score')
pit.show()
```

```python
x_range = np.arange(min(x),max(x),0.1)
x_range
```

```python
x_range=x_range.reshape(-1,1) # 2차원으로 만들기 row 개수는 자동으로 계산
```

```python
pit.scatter(x,y,color = 'blue')
pit.plot(x_range,lin_reg.predict(poly_reg.fit_transform(x_range)),color = 'green')
pit.title('score by hours')
pit.xlabel('hours')
pit.ylabel('score')
pit.show()
```

### 공부 시간에 따른 시험 성적 예측

```python
reg.predict([[2]])# 2시간 공부했을 때 선형 회귀 모듈의 예측
```

```python
lin_reg.predict(poly_reg.fit_transform([[2]]))
```

```python
lin_reg.score(x_poly,y)
```
